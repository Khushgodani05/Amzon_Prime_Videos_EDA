### **Important Observation:**
- We must handle the missing values of the titles dataset before merging both datasets.
- Becasue when we perform merging on common column "id" the credits dataset have multiple rows with same "id" which would dominate the handling of missing values in titles dataset.
- eg. if we want to fill age_criterion missing value with the mode of the column then the multiple rows with same "id" in credits dataset would affect the mode calculation of age_criterion column in titles dataset.
- Hence we will handle missing values in titles dataset first and then merge both datasets.

### Handling missing values

```python
#Here i am displaying the information about the titles dataset
data1.info()
```

```python
#Here since the number of missing values is very less for description and imdb_id we can drop those rows
data1=data1.dropna(subset=["description", "imdb_id"],axis=0)
data1.shape
```

```python
#Here i am displaying the count of unique values in age_certification column
data1.age_certification.value_counts().sort_values()
```

### Report:
- The 'R' age certification has the highest count among all age certifications and dominates the dataset.
- It depicts the majority of release were made for "R" rated contents.
- So we will fill its nan/missing values with the most frequent age_certification i.e 'R'.

```python
#Here i am displaying the statistical summary of titles dataset
data1.describe()
```

```python
# Here we perform the grouping by release year and finding mean , min and max of tmdb popularity to check if the release year has any effect on popularity
data1.groupby('release_year')['tmdb_popularity'].agg(["mean","min","max","median"]).sort_values(by="mean")
#we can see that the popularity is increasing with the release year
```

### Report 
- Here we have season value for shows and not for movie so we will fill the missing values with 0.
- Here we fill the age certification missing values with mode as it is categorical column.
- Here the mean and median of imdb_score is almost same so we will fill the missing value with mean.
- Here the mean and median of imdb_votes is not same so we will fill the missing value with median.
- Here the mean and median of tmdb_popularity is not same so we will fill the missing value with median.
- Here the mean and median of tmdb_score is almost same so we will fill the missing value with mean.
- Here we have very less missing values for description and imdb_id so we have drop those rows.

```python
#Here we are filling the missing values
from sklearn.impute import SimpleImputer
impute1=SimpleImputer(strategy="mean")
impute2=SimpleImputer(strategy="median")
impute3=SimpleImputer(strategy="most_frequent")
data1.seasons=data1.seasons.fillna(value=0)
data1.loc[:,["imdb_score","tmdb_score"]]=impute1.fit_transform(data1.loc[:,["imdb_score","tmdb_score"]])
data1.loc[:,["imdb_votes","tmdb_popularity"]]=impute2.fit_transform(data1.loc[:,["imdb_votes","tmdb_popularity"]])
data1.loc[:,["age_certification"]]=impute3.fit_transform(data1.loc[:,["age_certification"]])
```

INSIGHT: Amazon Prime maintains a movie-centric strategy with drama and comedy as dominant genres, serving primarily mature audiences while covering diverse categories from documentation to horror and romance.

2. REGIONAL AVAILABILITY - How does content distribution vary across different regions
---

KEY FINDINGS:
✓ Top 3 production countries:
  - United States: 4,535 titles (49% of catalog - clear dominance)
  - India: 955 titles (10% of catalog - second largest market)
  - Great Britain: 607 titles (6% of catalog - third largest)
✓ 100% production country data coverage (9,871 titles)
✓ US content represents better and more production quality compared to other countries

INSIGHT: US content heavily dominates the platform with nearly half the catalog, followed by significant representation from India and GB, indicating a primarily US-focused strategy with strategic international partnerships.

3. TRENDS OVER TIME - How has Amazon Prime's content library evolved
---

KEY FINDINGS:
✓ Majority of releases made after 2000 (modern content focus)
✓ Peak year: 2019 with 735 releases (highest single-year content addition)
✓ Consistent growth pattern showing library expansion over time
✓ Runtime distribution: Most content runs 80-120 minutes (standard feature film length)
✓ For shows: Majority have 1-10 seasons

INSIGHT: Content release on Amazon Prime has improved significantly over time, with 2019 marking the peak acquisition year. The platform shows consistent library growth with focus on standard-length features and multi-season shows.

4. IMDb RATINGS & POPULARITY - What are the highest-rated or most popular shows
---

KEY FINDINGS:

✓ Rating coverage statistics:
  - IMDb scores: 8,850 titles (90% coverage)
  - TMDb scores: 7,789 titles (79% coverage)
  - IMDb votes: 8,840 titles with engagement data

✓ Quality distribution:
  - IMDb scores: Maximum concentration between 5-7 range
  - TMDb scores: Maximum concentration between 5-7 range (consistent with IMDb)

✓ Engagement patterns:
